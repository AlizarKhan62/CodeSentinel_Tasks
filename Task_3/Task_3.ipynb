```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.head(3)
```

```python
df.info()
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['director'], inplace=True)  # Example: drop rows without director
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
df['country'].value_counts().head(5).plot(kind='bar', color='skyblue')
plt.title('Top 5 countries with most titles')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['release_year'],bins=30, kde=True)
plt.title('Distribution of release years')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='release_year', data=df)
plt.title('Boxplot of release years')
plt.show()
```

```python
df['type'].value_counts().plot(kind='bar')
plt.title('Movies vs TV shows')
plt.show()
```
